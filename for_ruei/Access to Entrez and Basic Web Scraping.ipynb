```python
from Bio import Entrez # 當我們要使用別人已經寫好的東西時候，除了安裝，就要先 `import`。
Entrez.email = "charlene50135@gmail.com" # 這裡只是再跟 ncbi 打個招呼
handle = Entrez.esummary(db="pubmed", id="14630660", retmode="xml") #Entrez的 esummary 這個功能可以幫你拿回一些簡單的資料
records = Entrez.parse(handle) # 只是他傳回來的是 XML，不好只取出我們的部份，所以要使用 parse （切割器）把他切成我們好去整理的樣子
for record in records: # 因為傳回來的東西，不一定只有一筆，所以我們要一筆一筆的印（print）出來看看有啥小
    print(record)
handle.close()# 事情做完記得要切斷連線
```

結果我發現 Entrez 能給的都沒有你要的，所以可能要從原始 journal 網站下手。要得到網站連結，就要從 doi 開始。隨便找個 doi 來看看好ㄌ。隨便用個 doi

```python
import requests #這些就是其他的包
from bs4 import BeautifulSoup

url = "http://dx.doi.org/10.7717/peerj.365"
news_page = requests.get(url) # 請把他的 html 傳回來
news = BeautifulSoup(news_page.text, 'html.parser') # 幫我把 html 切成好整理的樣子！
```

```python
news #發現在這裡面可以看到你想要的東西，有 corresponding
```

```python
news.find_all('link') # Beautiful soup 可以幫你找出所有叫做 link 的標籤，像是長這樣的東西：<link href="https://d2pdyyx74uypu5.cloudfront.net/" rel="dns-prefetch"/>
```

```python
import re 
re.findall(string = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}",news_page.text) # 想說先找出全部的文字中的 email，再去找出哪個標籤有那個 email，再往上推一層或往下推一層看誰是 corresponding
```

上面那一串`r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}"` 叫做 regular expression，他在想辦法表示 email 的 pattern 。詳細的寫法我也不懂，因為我是上網抄來的。有興趣可查查

```python
news.find_all(content = re.compile('dsolomon@msu.edu')) # 這會幫你找出 content 是那個 email 的標籤，可是沒有名字，很雞車
```

```python
news.find_all(href = re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}")) # 這可以找出 href (連結的意思)有 email 的標籤，可是還是沒有 author 名字
```

```python
news.find(title = re.compile("corresponding")).parent # 不然找找 title 有 corresponding 字串的好了，結果看到了我們要的
```
