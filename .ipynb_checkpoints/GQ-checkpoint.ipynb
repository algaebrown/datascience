```python
import pandas as pd

# read data
df = pd.read_excel('GQ.xlsx')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# set style
mpl.style.use("ggplot")

# plot inline
%matplotlib inline
```

```python
# assign appropriate data type
df['Start Date (UTC)'] = pd.to_datetime(df['Start Date (UTC)'])

import datetime as dt
# year of grad
def year_of_grad(datetime):
    
    if datetime.to_datetime() > dt.datetime(2017,1,1):
        year = 99
    else:
        year = 98
    return(year)
df['year'] = df['Start Date (UTC)'].apply(lambda x: year_of_grad(x))

df99 = df.groupby(['year']).get_group(99)
df98 = df.groupby(['year']).get_group(98)
```

```python
meanTotal = df.describe().loc['mean']
mean99 = df99.describe().loc['mean']
mean98 = df98.describe().loc['mean']
df_m = pd.concat([mean98, mean99, meanTotal], axis = 1, keys = ['98', '99', 'total'])
df_m['change'] = df_m['99']- df_m['98']
```

```python
fig = plt.figure(figsize = (5,50))
ax = fig.add_subplot(1, 2, 1)
plt.pcolor(df_m.iloc[0:110,0:3], cmap='gist_rainbow', vmin=1, vmax=5)
plt.title('整體表現綜覽：98級、99級分數平均值', fontsize = 'large')
plt.colorbar()
ax.set_xticks([0.5,1.5,2.5])
ax.set_xticklabels(['98(2016)', '99(2017)','total'], fontsize ='large', rotation = 30)
plt.gca().invert_yaxis()
ax.set_yticks(np.arange(0.5, 110.5, 1.0))
ax.set_yticklabels(list(df_m.index.values), fontsize ='large')

fig.text(0.46, 0.445, '平均分數', va='center', )

```

```python
# function

def group_plot(indexStart, indexEnd):
    # input: a group of questions, stored in `listOfAttr`
    # output: descriptive and plot, grouped by year
    indexList = list(range(indexStart, indexEnd))
    
    fig, axes = plt.subplots(nrows = indexEnd - indexStart, ncols = 4, figsize=(15, len(indexList)*3), sharex = True, sharey = True)
    rowInFig = 0
    
    fig.text(0.5, 0.07, '分數 1-5', ha='center', fontsize = 'large')
    fig.text(0.07, 0.5, '百分比', va='center', rotation='vertical', fontsize = 'large')
    
       
    
    for i in indexList:
        print('Q'+str(i)+':', df.columns.values[i])
        #fig.text(0.04, 1/len(indexList)*(rowInFig+1), 'Q'+str(i), ha='center')
        
        bin98 = axes[rowInFig,0].hist(df98.iloc[:,i], bins = 5, normed = True)
        axes[rowInFig,0].set_ylabel('Q'+str(i))
        
        bin99 = axes[rowInFig,1].hist(df99.iloc[:,i], bins = 5, normed = True, color = "#6666ff")
        axes[rowInFig,0].set_ylabel('Q'+str(i))
        
        binTotal = axes[rowInFig,2].hist(df.iloc[:,i], bins = 5, normed = True, color = "#808080")
        axes[rowInFig,0].set_ylabel('Q'+str(i))
        
        binChange = axes[rowInFig,3].hist([1,2,3,4,5], weights = tuple(np.array(bin99[0]) - np.array(bin98[0])), bins = 5, color = "#808080")
        axes[rowInFig,0].set_ylabel('Q'+str(i))
        rowInFig += 1 # next row in subplot
    
    axes[rowInFig-1,0].set_xlabel('98 級畢業生 (2016)')
    axes[rowInFig-1,1].set_xlabel('99 級畢業生 (2017)')
    axes[rowInFig-1,2].set_xlabel('所有畢業生')
    axes[rowInFig-1,3].set_xlabel('兩年差異（99級減去98級）')
    
    #fig.legend( (bin98, bin99, binTotal),('label1', 'label2', 'label3'))
    
    plt.savefig(str(rowInFig)+'GQ.png')    

    
```
