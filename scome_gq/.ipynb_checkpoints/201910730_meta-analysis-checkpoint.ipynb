```python
def plot_mean(df, start, end):
    
    df_m = df.iloc[start:end, :]
    fig = plt.figure(figsize = (5,(end-start)*0.5))
    ax = fig.add_subplot(1, 2, 1)
    plt.pcolor(df_m, cmap='coolwarm', vmin=1, vmax=5)
    plt.title('分數平均值', fontsize = 'large')
    plt.colorbar()
    ax.set_xticks(np.arange(0.5, df_m.shape[1], 1.0))
    ax.set_xticklabels(df.columns, fontsize ='large', rotation = 30)
    plt.gca().invert_yaxis()
    ax.set_yticks(np.arange(0.5, df_m.shape[0], 1.0))
    ax.set_yticklabels(list(df_m.index.values), fontsize ='large')

    fig.text(0.46, 0.445, '平均分數', va='center', )
    plt.savefig('../../gqfigs/{}.png'.format(str(df_m.index[0]).replace(':','')+str(df_m.columns[1]).replace(':','')), bbox_inches='tight', dpi = 300)
# 十分有趣的事情是：六年制在臨床實習的滿意度普遍比七年制高 (雖然沒有檢測 statistical significance)
```

```python
def plot_specialty(s):
    s.sort_values(by = s.columns[0], inplace = True)
    fig = plt.figure(figsize = (5,7))
    ax = fig.add_subplot(1, 2, 1)
    plt.pcolor(s, cmap='cividis', vmin=0, vmax=0.2)
    plt.title(s.index.name, fontsize = 'large')
    plt.colorbar()
    ax.set_xticks(np.arange(0.5, s.shape[1], 1.0))
    ax.set_xticklabels(s.columns.values, fontsize ='large', rotation = 30)
    plt.gca().invert_yaxis()
    ax.set_yticks(np.arange(0.5, s.shape[0], 1.0))
    ax.set_yticklabels(list(s.index.values), fontsize ='large')

    fig.text(0.46, 0.445, '比例', va='center', )
    plt.savefig('../../gqfigs/{}.png'.format(s.index.name + str(s.columns[1])), bbox_inches='tight', dpi = 300)
```

```python
specialty = df.loc[df['school']=='北醫'].groupby(by = 'year')['我計劃專科醫師選科的第一志願（不當醫師可跳過。沒有選項請選其他，並在三個志願後面填寫）'].value_counts(normalize = True).unstack().transpose()
plot_specialty(specialty)
```

```python
def plot_two_var(d):
    years = [98,99,101,102]
    colors = ['red', 'orange', 'green', 'blue']

    fig = plt.figure(figsize = (5,5))
    # list of markers for each hospital
    hp_label_dict = dict(zip(df_hp.columns.levels[1], ["1", "P", "v", "^", "*", "2","3","+","4","X","<",">","h","H"]))
    # list of colors for each year
    yr_label_dict = dict(zip(years, colors))

    dims = d.index
    for y in years:
         # hosptials of the year
        for h in d[y].columns:
            plt.scatter(x = d[y][h][dims[0]], y = d[y][h][dims[1]], color = yr_label_dict[y], marker = hp_label_dict[h], alpha = 0.6, label = str(y)+ str(h), vmin=1, vmax=5,)
    plt.xlabel(dims[0])
    plt.ylabel(dims[1])

    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.savefig('../../gqfigs/{}.png'.format(dims[0].split(':')[1]), bbox_inches='tight', dpi = 300)
```

```python
df_hp = df.loc[df['school']=='北醫'].groupby(by = ['實習醫院','year']).mean().transpose()
df_hp.columns.levels[0]
# sorted_list
hp_list = ['北萬雙','北醫', '萬芳', '雙和','台大', '北榮', '林長','馬偕','新光','國泰', '北慈', '奇美', '成大',   
        '高長']

# 科別
spec = [x.split('[')[1].split(']')[0] for x in df_hp.iloc[75:91].index] # 只有少數有多年 data
small_spec = ['內科', '外科','婦產科','小兒科']
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = [98,99,101,102]
colors = ['red', 'orange', 'green', 'blue']
yr_label_dict = dict(zip(years, colors))

def radar_plot(df_hospital):
    #標籤
    labels = np.array(['Supervision','指引回饋','學習意義','過夜原則'])

    # use length to determine angle
    dataLenth = len(labels)
    angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
    angles = np.concatenate((angles, [angles[0]])) # 閉合
    
    # initialize figure
    fig, axes = plt.subplots(nrows = len(hp_list), ncols = len(small_spec)
                             , subplot_kw=dict(polar=True)
                            , figsize = (5,18))
    #ax = fig.add_subplot(111, polar=True)# polar引數！！代表畫圓形！！！！
    #111代表總行數總列數位置
    
    def close_data(d):
        '''
        d: 一年份的資料 99, 北榮, 婦產 [4,4,5,5]
        '''

        #資料
        data = np.array(d)
        data = np.concatenate((data, [data[0]])) # 閉合 # #將資料結合起來
        
        return(data)

    
    def per_hp_spec(df_hospital, ax, hp_name, spec_index):
    
        # plot for every year
        for y in df_hospital.columns: 
            data = close_data(df_hospital[y])
            ax.plot(angles, data, 'bo-', linewidth=1, color = yr_label_dict[y], alpha = 0.6)# 畫線四個引數為x,y,標記和顏色，閒的寬度
        
        
        ax.set_thetagrids(angles * 180/np.pi, '', fontproperties="SimHei")
        ax.set_title(hp_name + ' ' + spec[spec_index], va='baseline', fontproperties="SimHei")
        ax.set_rlim(2,5)
        ax.grid(True)
        
        return(ax)
            
    
    def spec_loop(hp_name, hp_index):
        
        for i in range(len(small_spec)): # loop over specialty
            # look back to the original all speciality list for the true index in df
            # i for axes location
            spec_index = spec.index(small_spec[i])
            # extract dataframe
            
            df_hospital = df_hp.iloc[[75 + spec_index, 75 + 16 + spec_index, 75 + 16*2 + spec_index, 75 + 16*3 + spec_index]][hp_name]
            
            per_hp_spec(df_hospital, axes[hp_index,  i], hp_name, spec_index)
    
    for hp_index in range(len(hp_list)):
        spec_loop(hp_list[hp_index], hp_index)    
    plt.show()
    
    return(plt)
plt = radar_plot(df_hp)
plt.savefig('../../gqfigs/hp_spec.png', bbox_inches='tight', dpi = 300)
```

```python
# legend for radar
def radar_plot_legend(df_hospital):
    #標籤
    labels = np.array(['Supervision','指引回饋','學習意義','過夜原則'])

    # use length to determine angle
    dataLenth = len(labels)
    angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
    angles = np.concatenate((angles, [angles[0]])) # 閉合
    
    # initialize figure
    fig = plt.figure(figsize = (9,9))
    ax = fig.add_subplot(121, polar=True)# polar引數！！代表畫圓形！！！！
    #111代表總行數總列數位置
    
    def close_data(d):
        '''
        d: 一年份的資料 99, 北榮, 婦產 [4,4,5,5]
        '''

        #資料
        data = np.array(d)
        data = np.concatenate((data, [data[0]])) # 閉合 # #將資料結合起來
        
        return(data)
    # plot for every year
    for y in df_hospital.columns: 
        data = close_data(df_hospital[y])
        ax.plot(angles, data, 'bo-', linewidth=1, color = yr_label_dict[y], alpha = 0.6, label = y)# 畫線四個引數為x,y,標記和顏色，閒的寬度
        
        
    ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
    ax.set_title('OO醫院 OO科', fontproperties="SimHei")
    ax.set_rlim(2,5)
    ax.grid(True)
    
    ax.legend(bbox_to_anchor=(1.1, 1.05))
     
```

```python
d = df_hp.iloc[[75,75 + 16 , 75 + 16*2, 75 + 16*3]]['北榮']
d[102] = [4.9,3,5,5]

radar_plot_legend(d)
```
