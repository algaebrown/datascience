```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time

# this opens a new Firefox window!
# make sure that you have geckodriver installed and have it added to the $PATH
# https://stackoverflow.com/questions/40208051/selenium-using-python-geckodriver-executable-needs-to-be-in-path
driver = webdriver.Firefox()

# maximize the window
driver.maximize_window()

# now let'sgo striaght to our target!
url = 'http://wwwc.moex.gov.tw/main/exam/wFrmExamQandASearch.aspx?menu_id=241' # ministry of examination
driver.get(url) 

# getting the title and url: to confirm that our action is getting what we want
print("this is title!", driver.title)
print("current url", driver.current_url)

### go find our target: we want all links to old questions and its answer

# now what we see is "簡易查詢" and we need to switch to "切換至進階查詢"
# its id: ctl00_holderContent_ibtnFull
# find by its id yee and CLICK without hesitation yee
driver.find_element_by_id('ctl00_holderContent_ibtnFull').click()
# show off the result
print("this is title!", driver.title)

# then we need to select "考試年度"
# its id: ctl00_holderContent_wUctlExamYearStart_ddlExamYear ctl00_holderContent_wUctlExamYearEnd_ddlExamYear
# example: start(106) to End(81)
# drop-down list: https://jzchangmark.wordpress.com/2015/03/05/%E9%80%8F%E9%81%8E-selenium-%E6%93%8D%E4%BD%9C%E4%B8%8B%E6%8B%89%E5%BC%8F%E9%81%B8%E5%96%AE-select/
select = Select(driver.find_element_by_id('ctl00_holderContent_wUctlExamYearStart_ddlExamYear'))
select.select_by_visible_text(u"106")
select = Select(driver.find_element_by_id('ctl00_holderContent_wUctlExamYearEnd_ddlExamYear'))
select.select_by_visible_text(u"81")

# fill in "類科"
# its id: ctl00_holderContent_txtCategoryName
element = driver.find_element_by_id("ctl00_holderContent_txtCategoryName")
element.send_keys(u"醫師（二）")

# press submit "查詢"
# its id: ctl00_holderContent_btnQuery
driver.find_element_by_id('ctl00_holderContent_btnQuery').click()

#click on 試題
Questions = driver.find_elements_by_xpath("//*[contains(text(), '試題')]")
print(Questions)

# this is to suspend the process so that we can see our window. after three secs the window quits due to the following `quit` command
time.sleep(3) 

# quit the driver is the good habit!! 
driver.quit()
```
